# Education Data API Exploration

## Objective
Demonstrate ability to:
- Make REST API calls
- Apply filters
- Parse JSON responses
- Convert data to structured format
- Prepare data for reporting

## Endpoints Used
- CCD Directory 2013
- CCD Enrollment Grade 3 (Charter, DC filter)

```python
import requests
import json

url = "https://educationdata.urban.org/api/v1/schools/ccd/directory/2013/"

response = requests.get(url)

print("Status code:", response.status_code)

data=response.json()

print(json.dumps(data, indent=2) [:2000])

#using REST endpoint, pasing JSON, validating status code

```

```python
url = "https://educationdata.urban.org/api/v1/schools/ccd/enrollment/2013/grade-3/?charter=1&fips=11"

response = requests.get(url)
data = response.json()

print("Number of Records:", len(data["results"]))
print(json.dumps(data["results"][:2], indent=2))
```
